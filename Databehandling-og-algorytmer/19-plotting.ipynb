```python
import json 
import matplotlib.pyplot as plt 

with open("land.json", encoding="utf-8") as fil:
    data = json.load(fil)

data[0]
```

```python
verdensdeler = {}
for land in data:
    print(land['verdensdel'])
    if land ['verdensdel'] not in verdensdeler:
        print("fant en ny verdensdel! putter den i ordboka")
        verdensdeler[land['verdensdel']] = 1
    else:
        verdensdeler[land['verdensdel']] += 1

print(verdensdeler)
```

```python
land = list(verdensdeler.keys())
antall = list(verdensdeler.values())

plt.bar(land, antall)
plt.title("Antall land på topplista ")
plt.xlabel("verdensdeler")
plt.ylabel("antall")
plt.style.use("fivethirtyeight")
```
